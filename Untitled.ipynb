```python
import sys
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('7.png')
grayImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)#转换成单通道图像

cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",grayImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
guassian = cv2.GaussianBlur(grayImg, (3,3), 0)

cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",guassian)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(guassian)
plt.show()
guassian
```

```python
sobelx = cv2.Sobel(guassian, cv2.CV_16S, 1, 0, (3,3))
sobely = cv2.Sobel(guassian, cv2.CV_16S, 0, 1, (3,3))





# cv2.namedWindow("srcx1", cv2.WINDOW_NORMAL)
# cv2.imshow("srcx1",sobelx)

# cv2.namedWindow("srcy1", cv2.WINDOW_NORMAL)
# cv2.imshow("srcy1",sobely)


sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)

sobel = sobelx - sobely

# cv2.namedWindow("srcx", cv2.WINDOW_NORMAL)
# cv2.imshow("srcx",sobelx)

# cv2.namedWindow("srcy", cv2.WINDOW_NORMAL)
# cv2.imshow("srcy",sobely)



cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
ret,erzhihua = cv2.threshold(sobel, 0 , 255, cv2.THRESH_OTSU)

cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",erzhihua)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(erzhihua)
plt.show()
```

```python
element = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))
element3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
element5 = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))

bi = cv2.morphologyEx(erzhihua, cv2.MORPH_CLOSE, element3)

# bi = cv2.morphologyEx(bi, cv2.MORPH_ERODE, element)
# #bi = cv2.morphologyEx(bi, cv2.MORPH_ERODE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_ERODE, element5)

# #bi = cv2.morphologyEx(bi, cv2.MORPH_CLOSE, element)

# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)
# bi = cv2.morphologyEx(bi, cv2.MORPH_DILATE, element)


bi = cv2.morphologyEx(bi, cv2.MORPH_CLOSE, element)


cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",bi)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contours, hierarchy = cv2.findContours(bi, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
img2 = cv2.drawContours(img, contours, -1, (0,255,0), 3)
cv2.namedWindow("src", cv2.WINDOW_NORMAL)
cv2.imshow("src",img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
